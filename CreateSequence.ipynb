```python
import nltk
import numpy as np
ConvertedData = []
for tweet in data:
    t = np.where(np.array(tweet) != '')
    #get the tags
    tagged = np.array(nltk.pos_tag(tweet[:len(t[0])]))[:,1]
    #convet the tags to numbers
    tagsN = [d[tag] for tag in tagged]
    
    #put in a zeroed list
    zeroed = np.zeros(70)
    zeroed[:len(tagsN)]= tagsN
    onehotEncoded = (np.arange(47) == zeroed[...,None]-1).astype(int)
    ConvertedData.append(onehotEncoded)
    
ConvertedData = np.array(ConvertedData)
print(ConvertedData)  
```

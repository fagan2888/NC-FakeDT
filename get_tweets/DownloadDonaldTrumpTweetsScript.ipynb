```python
def get_tweets(username):

    #http://tweepy.readthedocs.org/en/v3.1.0/getting_started.html#api
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_key, access_secret)
    api = tweepy.API(auth)

    #set count to however many tweets you want; twitter only allows 200 at once
    processedTweets = []


    #load the ids
    with open('all_ids.json') as f:
        ids = json.load(f)

    print('total ids: {}'.format(len(ids)))

    start = 0
    end = 100
    limit = len(ids)
    i = math.ceil(limit / 100)

    for go in range(i):
        print('currently getting {} - {}'.format(start, end))
        sleep(6)  # needed to prevent hitting API rate limit
        id_batch = ids[start:end]
        start += 100
        end += 100
        
        tweets = []
        count = 0
        for instance in id_batch:
            count+=1
            count_min = 0
            while True:
                try:
                    
                    tweets.append(api.get_status(instance, tweet_mode = "extended"))
                    print("counter at {} with foud size {}".format(count+start-100, len(tweets)))
                    break
                
                except tweepy.TweepError:
                    print ("Could not get tweet at counter {}, waiting since {} min".format(count+start-100, count_min))
                    print ("wait 1 min")


                    for i in tqdm(range(60)):
                         time.sleep(1)
                    
                    #sleep(60)
                    count_min +=1
            
        #tweets = api.statuses_lookup(id_batch, )
        
 #       print(tweets)

        # preprocess the tweets
        for i in range(len(tweets)):

            indices = []
            tweet = tweets[i]
            text = tweet.full_text

            # dont add retweets
            if 'RT @'  not in text:

                urls = tweet.entities["urls"]
                for url in urls:
                    indices.append(url["indices"])

                if "media" in tweet.entities.keys():
                    urls = tweet.entities["media"]
                    for url in urls:
                        indices.append(url["indices"])

                processedTweet = [username, ""]
                # remove urls
                if len(indices) == 0:
                    processedTweet[1] = text 
                else:
                    sorted(indices, key = itemgetter(0))
                    pos = 0

                    for index in indices:
                        processedTweet[1] = processedTweet[1] + text[pos:index[0]]  
                        pos = index[1]
                    processedTweet[1] = processedTweet[1] + text[pos:len(text)]


                # encode to csv understandable format    
                processedTweet[1] = processedTweet[1].encode("utf-8")
                processedTweets.append(processedTweet)


    # Print to csv file
    print ("writing to {0}_tweets.csv".format(username))

    with open("{0}_tweets.csv".format(username) , 'w+') as file:
        writer = csv.writer(file, delimiter='|')
        writer.writerows(processedTweets)
        
```
